# 1. Descargando datos a la fuerza bruta

Vamos a ver primero como ir descargando datos y luego como lidiar con diferentes formatos. Es muy importante que, si los datos los vamos a cargar por única vez, descargar el conjunto de datos, tal como se encuentran, esto es `raw data`.

Vamos primero cargando las bibliotecas necesarias:

```python

import os  # Para manejo de archivos y directorios
import urllib.request # Una forma estandard de descargar datos
# import requests # Otra forma no de las librerías de uso comun

import datetime # Fecha de descarga
import pandas as pd # Solo para ver el archivo descargado
import zipfile # Descompresión de archivos
```

Es importante saber en donde nos encontramos y crear los subdirectorios necesarios para guardar los datos de manera ordenada. Tambien es importante evitar cargar datos que ya han sido descargados anteriormente.

```python
# pwd
print(os.getcwd())

#  Estos son los datos que vamos a descargar y donde vamos a guardarlos
desaparecidos_RNPDNO_url = "http://www.datamx.io/dataset/fdd2ca20-ee70-4a31-9bdf-823f3c1307a2/resource/d352810c-a22e-4d72-bb3b-33c742c799dd/download/desaparecidos3ago.zip"
desaparecidos_RNPDNO_archivo = "desaparecidosRNPDNO.zip"
desaparecidos_corte_nacional_url = "http://www.datamx.io/dataset/fdd2ca20-ee70-4a31-9bdf-823f3c1307a2/resource/4865e244-cf59-4d39-b863-96ed7f45cc70/download/nacional.json"
desaparecidos_corte_nacional_archivo = "desaparecidos_nacional.json"
subdir = "./data/"
```

```python
if not os.path.exists(desaparecidos_RNPDNO_archivo):
    if not os.path.exists(subdir):
        os.makedirs(subdir)
    urllib.request.urlretrieve(desaparecidos_RNPDNO_url, subdir + desaparecidos_RNPDNO_archivo)
    with zipfile.ZipFile(subdir + desaparecidos_RNPDNO_archivo, "r") as zip_ref:
        zip_ref.extractall(subdir)

    urllib.request.urlretrieve(desaparecidos_corte_nacional_url, subdir + desaparecidos_corte_nacional_archivo)

    with open(subdir + "info.txt", 'w') as f:
        f.write("Archivos sobre personas desaparecidas\n")
        info = """
        Datos de desaparecidos, corte nacional y desagregación a nivel estatal,
        por edad, por sexo, por nacionalidad, por año de desaparición y por mes
        de desaparición para los últimos 12 meses.

        Los datos se obtuvieron del RNPDNO con fecha de 03 de agosto de 2021
        (la base de datos no se ha actualizado últimamente)

        """
        f.write(info + '\n')
        f.write("Descargado el " + datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S") + "\n")
        f.write("Desde: " + desaparecidos_RNPDNO_url + "\n")
        f.write("Nombre: " + desaparecidos_RNPDNO_archivo + "\n")
        f.write("Agregados nacionales descargados desde: " + desaparecidos_corte_nacional_url + "\n")
        f.write("Nombre: " + desaparecidos_corte_nacional_archivo + "\n")
```

Utiliza los archivos `json` descargados con el detalle a nivel estatal, y genera unos 3 `DataFrame` con información sobre personas desaparecidas dependiendo de diferentes características.

```python
df_desaparecidos = pd.read_json(subdir + desaparecidos_corte_nacional_archivo)
df_desaparecidos
```

Como primer DataFrame del ejercicio, se generara uno que permita identificar cuantos desaparecidos por sexo hay en cada estado.

```python
hombres_desaparecidos = df_desaparecidos["espacial"].loc["Hombre"] # Extraemos el diccionario (o json anidado) correspondiente a hombres
mujeres_desaparecidas = df_desaparecidos["espacial"].loc["Mujer"] # Repetimos con mujeres
indet_desaparecidos = df_desaparecidos["espacial"].loc["Indeterminado"] # Y de igual manera con Indeterminados
df_1 = pd.DataFrame.from_dict(hombres_desaparecidos, orient="index", columns=["hombres"]) #Generamos un Dataframe desde el primer diccionario
df_1["mujeres"] = df_1.index.map(mujeres_desaparecidas) # Y hacemos uso de la funcion map para agregar el diccionario de mujeres al dataframe
df_1["indeterminados"] = df_1.index.map(indet_desaparecidos) # Repetimos con los intedeterminados
df_1
```

Como segundo DataFrame del ejercicio, se generara uno que permita identificar cuantos desaparecidos por edad hay en cada estado.

```python
hombres_desaparecidos = df_desaparecidos["por_edad"].loc["Hombres"] # Extraemos el diccionario (o json anidado) correspondiente a hombres
mujeres_desaparecidas = df_desaparecidos["por_edad"].loc["Mujeres"] # Repetimos con mujeres
indet_desaparecidos = df_desaparecidos["por_edad"].loc["Indeterminado"] # Y de igual manera con Indeterminados
df_2 = pd.DataFrame.from_dict(hombres_desaparecidos, orient="index", columns=["hombres"]) #Generamos un Dataframe desde el primer diccionario
df_2["mujeres"] = df_2.index.map(mujeres_desaparecidas) # Y hacemos uso de la funcion map para agregar el diccionario de mujeres al dataframe
df_2["indeterminados"] = df_2.index.map(indet_desaparecidos) # Repetimos con los intedeterminados
df_2 = df_2.sort_index() # Efectuamos un acomodo de los indices (strings debido al valor de SIN EDAD ASIGNADA)
df_2
```

Finalmente, para el tercer dataframe, buscaremos captar por nacionalidad que sexo sufrio mayor incidencia de desaparicion.

```python
hombres_desaparecidos = df_desaparecidos["por_nacionalidad"].loc["Hombres"] # Extraemos el diccionario (o json anidado) correspondiente a hombres
mujeres_desaparecidas = df_desaparecidos["por_nacionalidad"].loc["Mujeres"] # Repetimos con mujeres
df_3 = pd.DataFrame.from_dict(hombres_desaparecidos, orient="index", columns=["hombres"]) #Generamos un Dataframe desde el primer diccionario
df_3["mujeres"] = df_3.index.map(mujeres_desaparecidas) # Y hacemos uso de la funcion map para agregar el diccionario de mujeres al dataframe
df_3["sexo_mayor_incidencia"] = df_3["hombres"]>df_3["mujeres"] # Generamos la columna que utiliza condicion para determinar el sexo con mas desapariciones
df_3["sexo_mayor_incidencia"] = ["hombre" if x else "mujer" for x in df_3["sexo_mayor_incidencia"]] # Reemplazamos valores booleanos con categorias (str)
df_3
```

# 3. Archivos xml

Los archivos *xml* son una manera de compartir información a través de internet o de guardar información con formatos genéricos que sigue siendo muy utilizada hoy en día. En general lidiar con archivos xml es una pesadilla y se necesita explorarlos con calma y revisarlos bien antes de usarlos.

La definición del formato y su uso se puede revisar en [este tutorial de la w3schools](https://www.w3schools.com/xml/default.asp). Vamos a ver un ejemplo sencillo basado en la librería [xml.etree.ElementTree](https://docs.python.org/3/library/xml.etree.elementtree.html) que viene de base en python:


```python
import xml.etree.ElementTree as et

archivo_url = "https://github.com/mcd-unison/ing-caract/raw/main/ejemplos/integracion/ejemplos/ejemplo.xml"
archivo_nombre = "ejemplito.xml"
subdir = "./data/"

if not os.path.exists(subdir + archivo_nombre):
    if not os.path.exists(subdir):
        os.makedirs(subdir)
    urllib.request.urlretrieve(archivo_url, subdir + archivo_nombre)


desayunos = et.parse(subdir + archivo_nombre)

for (i, des) in enumerate(desayunos.getroot()):
    print("Opción {}:".format(i+1))
    for prop in des:
        print("\t{}: {}".format(prop.tag, prop.text.strip()))

# Se puede buscar por etiquetas y subetiquetas

print("Los desayunos disponibles son: " +
      ", ".join([p.text for p in desayunos.findall("food/name")]))

# ¿Como se podría poner esta información en un DataFrame de `pandas`?
# Agreguen tanto código como consideren necesario.
```

```python
ns = "http://www.mediawiki.org/xml/export-0.10/" # Asignamos el namespace a una variable tipo string para utilizarlo en pasos subsecuentes
df_poetas_xml = pd.DataFrame() # Generamos un DataFrame vacio para llenarlo en un ciclo "for" en la siguiente celda
df_poetas_xml
```

El archivo XML tiene una estructura de arbol, donde cada elemento tiene sub-elementos, o como suelen llamarse, childs. En este caso tenemos que acceder a los childs desde un namespace definido en la raiz de mediawiki o de lo contrario no podremos visualizar atributos como el titulo de manera correcta. Estos archivos XML son bastante extensos, por lo que no se recomienda imprimir la informacion desde ciclos anidados, si no explorarla por partes y en el codigo depositar directamente la informacion a un formato de otro tipo, como el tabular.

```python
nombres_poetas = []

for page in root.findall('{%s}page' % ns): # Generamos un ciclo donde llenaremos la columna de nombres de nuestro dataframe
  nombres_poetas.append(page.find('{%s}title' % ns).text)

df_poetas_xml["nombres_poetas"] = nombres_poetas
df_poetas_xml
```
